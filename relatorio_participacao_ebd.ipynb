```python
import matplotlib.pyplot as plt

semana = ['Primeira', 'Segunda', 'Terceira', 'Quarta', 'Quinta', 'Sexta']
participacao = [35, 34, 29, 34, 35, 31]

plt.figure(figsize=(10, 6))
plt.style.use('fivethirtyeight')
barras =plt.bar(semana, participacao)
plt.bar_label(barras, labels= participacao, fontsize=12, color='#f09580', fontweight='bold')
plt.xlabel('Semanas',fontweight='bold', fontsize=14)
plt.ylabel('Participação',fontweight='bold', fontsize=14)   
plt.yticks([])
plt.title('Participação Semanal da EBD', fontsize=18, fontweight='bold', color='gray')
plt.xticks(rotation=45)
plt.box(False)
plt.grid(False)

plt.savefig('grafico_participacao.png', bbox_inches='tight', dpi=120)

plt.show()

```

```python
media_participacao = sum(participacao) / len(participacao)
print(f"Média de participação: {media_participacao:.2f}")
```

```python
total_membros = 47
porcentagem_participacao = (media_participacao / total_membros) * 100
print(f"Porcentagem média de participação: {porcentagem_participacao:.2f}%")
```
